```python
import matplotlib.pyplot as plt
import numpy as np
from random import randint as rnd
```

```python
plt.xlabel("Yıllar")
plt.ylabel("İhracat")
plt.title("Yıllara göre ihracat")

xpoint = np.array([i for i in range(2010,2016)])
ypoint = np.array([rnd(800,1200) for i in range(6)])
plt.plot(xpoint,ypoint,marker='D',ms=15,mec='r',mfc='g',linestyle="-.")
plt.show()
```

```python
font1={"family":"arial","color":"red","size":10}
font2={"family":"serif","color":"darkblue","size":15}

plt.title("Accuracy Graph",fontdict=font1,loc="left")
plt.xlabel("Epoch",fontdict=font2)
plt.ylabel("Loss",fontdict=font2)
plt.grid()
yTrainPoint = np.array([0.8,0.75,0.6,0.15,0.3])
yValPoint = np.array([0.8,0.65,0.6,0.5,0.6])

plt.plot(yTrainPoint,linestyle=':',c='red',lw=2,marker='X')
plt.plot(yValPoint,linestyle='--',c='green',lw=2,marker='D')
plt.legend("Train Loss","Val Loss",loc=0)
plt.show()
```

```python
y1TrainPoint = np.array([0.8,0.75,0.6,0.15,0.3])
plt.subplot(1,3,1)
plt.plot(y1TrainPoint)
plt.title("Train Graph")
plt.xlabel("Epoch")
plt.ylabel("Loss")

y1ValPoint = np.array([0.55,0.45,0.3,0.35,0.4])
plt.subplot(1,3,2)
plt.plot(y1TrainPoint)
plt.title("Val Graph")
plt.xlabel("Epoch")
plt.ylabel("Loss")

y1TestPoint = np.array([0.55,0.45,0.3,0.35,0.4])
plt.subplot(1,3,3)
plt.plot(y1TrainPoint)
plt.title("Test Graph")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.show()
```

```python
x = np.array([rnd(1,100) for i in range(100)])
y = np.array([rnd(1,100) for i in range(100)])
colors = np.array([rnd(1,100) for i in range(100)])
sizes = 10 * np.array([rnd(1,100) for i in range(100)])

plt.scatter(x,y,c=colors,s=sizes, alpha=0.5,cmap='gist_rainbow')
plt.colorbar()
plt.show()
```

```python
plt.title("Yılık Meyve Tüketimi (TON)")
x1 = np.array(['Elma','Armut','Çilek','Muz','Şeftali'])
y1 = np.array([100,30,90,60,10])

plt.barh(x1,y1,color=['red','yellow','green','pink','blue'],height=0.5)
plt.show()
```

```python
y=np.array([rnd(1,100) for i in range(1000)])
x=np.array([i for i in range(1,102)])
plt.title("Histogram")

plt.hist(y,x,color='green')
plt.show()
```

```python
pie = np.array([25,30,15,5,90])
label = np.array(["Toyota","hyundai","Ferrari","Volvo","Mercedes"])
plt.pie(pie,labels=label,colors=['yellow',"green","blue","pink","red"],autopct="%1.1f",
        explode=[0,0.6,0,0,0],shadow=True)

plt.legend(title="------Arabalar------",loc='upper left')
plt.show()
```
