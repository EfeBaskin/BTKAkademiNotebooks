```python
def working_year(year):
    return int(year) - 2006

# Apply the function to the 'JoiningYear' column
data['JoiningYear'] = data['JoiningYear'].apply(working_year)
```

```python
data['Education'].unique(),data['Gender'].unique(),data['City'].unique()
```

```python
data['Education'] = data['Education'].map({'Bachelors':0,'Masters':1,'PHD':2})
data['Gender'] = data['Gender'].map({'Male':0,'Female':1})
data['City'] = data['City'].map({'Bangalore':0,'Pune':1,'New Delhi':2})
```

```python
data.head()
```

```python
data['Age'].unique()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Normalize the 'Age' column
data['Age'] = scaler.fit_transform(data[['Age']])
data['JoiningYear'] = scaler.fit_transform(data[['JoiningYear']])
data['ExperienceInCurrentDomain'] = scaler.fit_transform(data[['ExperienceInCurrentDomain']])
```

```python
data.head()
```

```python
x = data.iloc[:,0:8]
y = data.LeaveOrNot
```

```python
x,x.shape
```

```python
y
```

```python
x.drop(['ExperienceInCurrentDomain',"EverBenched"],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=21)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

def largest(arr):
    mx = arr[0]
    for i in range(1,len(arr)-1):
        if arr[i] > mx:
            mx = arr[i]
    return mx

for i in range(1,11):
  arr=[]
  model = RandomForestClassifier(n_estimators=i)
  model.fit(x_train, y_train)
  y_test_pred = model.predict(x_test)
  val=accuracy_score(y_test,y_test_pred)
  print(val)
  arr.append(val)

if largest(arr):
  print(f"best estimator for model is {i}")
  model = RandomForestClassifier(n_estimators=i)
  model.fit(x_train, y_train)
  y_test_pred = model.predict(x_test)
  print(classification_report(y_test,y_test_pred))
  print(confusion_matrix(y_test,y_test_pred))
```
