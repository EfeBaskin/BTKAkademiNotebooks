```python
# ----Veri Seti Bölümlendirme İşlemi----
# -Holdout -Cross-Validation
# -parametre ayarlamaları için validation kullanılmaz

# ----overfitting önlemek için Tavsiyeler----
# -Modeli sadeleştir
# -daha az feature ekle (feature selection)
# -Daha fazla örnek ekle
# -Data augmentation ile veri çoğalt
# -Düzleştirme(Regularization) yap
# -erken durdurma algoritmaları kullan
# -Cross Validation kullan
# -Bunlar olmuyorsa farklı bir model kullan

# ----Underfitting önlemek için tavsiyeler----
# Data augmentation ve sentetik veri öğretimi tekniklerini kullan
# Daha fazla feature ekle
# Daha fazla örnek ekle
# Daha uzun süre train yap
# Olmuyorsa başka model dene

# ----Cross Validation----
# Dataset n bölümr ayrılır
# Veri Seti foldlara bölünerek verilen orana göre accuracy hesaplanır
# En son ortalamaları alınır ve accuracy değerine ulaşılır

# ----Feature Selection----
# Filtre , Örtü , Göümülü 3 çeşidi vardır
# Overfittinge sebep olabilmektedir

# ----Standardizasyon(z-score)----
# ----Normalizasyon (min-max normalizasyon)
# -> min 0 max 1 olacak şekilde verileri normalize etmek

# Sınıflandırmada performans değerlendirme
# Accuracy , Precision , F1-score , ROC , AUC
# Confusion Matrix
# -> Eğer veri seti unbalanced veya balanced ise
# -> Accuracy her zaman doğru sonuç vermez
# -> veri seti balanced olmak zorunda
# Accuracy = TP+TN / TP+FP+TN+FN (confusion matrix)
# Precision = TP/TP+FP
# Recall = TP/TP+FN Recall oldukça yüksek olmalı
# F1-Score = 2*Recall*Precision/Recall+Precision
# ROC ve AUC imbalanced datasetler için kullanılan yöntemlerdir
```

```python
# Örnek
# tp 100  fp 150
# fn 300  tn 4450

# Accuracy 4550/5000 = 0.91
# Recall 100/400 = 0.25
# Precision 100/250 = 0.4
# F1-Score 2*1/4*2/5/0.65 = 0.307
```
