```python
import pandas as pd
```

```python
dataset = pd.read_csv(r'/content/drive/MyDrive/heart.csv')
dataset
```

```python
dataset.info()
```

```python
bins = [20,40,60,100]
labels = ["20-40","40-60","60-100"]

dataset["age"] = pd.cut(dataset["age"],bins=bins,labels=labels,right=False)
dataset
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
dataset[["trtbps","chol","thalachh"]] = scaler.fit_transform(dataset[["trtbps","chol","thalachh"]])
```

```python
dataset
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
dataset["age"] = le.fit_transform(dataset["age"])
```

```python
dataset
```

```python
Xsample = dataset.drop("output",axis=1)
Ysample = dataset["output"]
```

```python
Xsample, Ysample
```

```python
dataset.shape
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(Xsample,Ysample,test_size=0.2,random_state=42)
x_train,x_val,y_train,y_val = train_test_split(x_train,y_train,test_size=0.25,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(x_val.shape)
print(y_train.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix

model=RandomForestClassifier()
model.fit(x_train,y_train)

y_test_pred = model.predict(x_test)
print(classification_report(y_test,y_test_pred))
print(confusion_matrix(y_test,y_test_pred))
```
