**Genel CNN Mimarisi**
---
* Convolutional Layer -> Özellik çıkarma için kullanılır
* Non-Linearity -> Sisteme doğrusal olmayanlığın tanıtılması , aktivasyon fonksşyonları
- Pooling (Downsampling ) Layer -> Ağırlık sayısını azaltır ve uygunluğu kontrol eder
* Flattening Layer -> Klasik sinir ağı için verileri hazırlar
* Fully Connected Layer -> sınıflamada kullanılan standart sinir ağı


**Relu formula = F(x)=max(0,x)**
* Katman aktivasyon katmanı olarak tanımlanır.
* Yapay sinir ağlarında kullanılan tanh,sigmoid,relu vb. aktivasyon fonksiyonları bu katmanda kullanılmaktadır.
* Performans olarak diğerlerinden daha hızlı çalışır.

**Pooling Layer**
---
* Amacı convolution layerdan sonra oluşan tabloyu özgün bir şekilde minimize etmektir.
* Gereksiz özellikler yok sayılır ve işlem gücünden tasarruf edilir.
* En çok kullanılan max poolingtir.
* Max pooling dışında ; average pooling, min pooling gibi çeşitleri de vardır.
* Convolution layerdan sonra sıkla eklenen bir katmandır.


**Flatten Layer**
---
* Matrisin vektörleştiği katmandır
* Görevi sonraki katman olan fully connect yapay sinir ağına işleyebileceği veriler oluşturmaktır
* Tek boyutlu dizilerle çalışır
* Bu katmanda matris olarak gelen verileri düzleştirerek vektör haline getirip bir sonraki katmana hazır hale getirir


**Fully Connect Layer**
---
* Mimarinin son bölümüdür
* Bu bölüm klasik yapay sinir ağından oluşmuştur
* En son layer classification yapılan kısımdır
* Dropout ve batchnormalization fully connect layer'da uygulanmaktadır
* Genellikle son aşamasında aktivasyon fonksiyonu olarak softmax uygulanmaktadır



**Dropout**
* Overfitting olma durumunda bazı bağlantılar unutulup ağ yeniden eğitilir
* Fine tuning için 0.5'ten başlayarak yavaş yavaş değer azaltılabilir
* Sistem stabil olduğunda değer azaltması durdurulur

**Batch Normalization**
---
* Normalde normalize edilen veriler convolution sürecinde dağılım bozukluğuna uğrayabilir.Nu eğitim hızını ve kararlılığını etkileyebilir.
* Fully connect bölümünde batch normalization işlemi ile veriler yeniden dengeli hale getirilebilir.
* Kullanıldığında lossta düşüş meydana gelir.
